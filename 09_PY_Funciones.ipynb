## ¿Por qué usar Funciones?

Las funciones son **bloques de código reutilizables**. En lugar de escribir el mismo código varias veces, lo defines una vez en una función y lo llamas cuando lo necesites.

**Beneficios:**
- **Reutilización**: escribe una vez, usa muchas veces
- **Organización**: código más limpio y legible
- **Mantenimiento**: si hay un error, lo corriges en un solo lugar

## Definir una función

```python
def nombre_funcion(parametro1, parametro2):
    """Descripción de lo que hace la función."""
    # cuerpo de la función
    resultado = ...
    return resultado
```

````python
# Función simple: saludar
def saludar(nombre):
    """Genera un saludo personalizado."""
    mensaje = f'¡Hola, {nombre}! Bienvenido al sistema.'
    return mensaje

# Llamar (invocar) la función
print(saludar('Ana'))
print(saludar('Carlos'))
````

````python
# Función con múltiples parámetros y retorno
def calcular_precio_con_iva(precio, tasa_iva=0.15):
    """Calcula el precio con IVA incluido.

    Args:
        precio: Precio base del producto
        tasa_iva: Tasa del IVA (por defecto 15%)

    Returns:
        Precio total con IVA
    """
    iva = precio * tasa_iva
    total = precio + iva
    return total

# Usar la función
print(f'Precio con IVA 15%: ${calcular_precio_con_iva(100):.2f}')
print(f'Precio con IVA 12%: ${calcular_precio_con_iva(100, 0.12):.2f}')
````

````python
# Función que retorna múltiples valores (como una tupla)
def estadisticas_ventas(datos):
    """Calcula estadísticas básicas de una lista de ventas."""
    total   = sum(datos)
    promedio = total / len(datos)
    maximo  = max(datos)
    minimo  = min(datos)
    return total, promedio, maximo, minimo

ventas = [3200, 4100, 2900, 5600, 3800, 4500]
total, prom, maximo, minimo = estadisticas_ventas(ventas)

print('=== ESTADÍSTICAS DE VENTAS ===')
print(f'Total:    ${total:,}')
print(f'Promedio: ${prom:,.2f}')
print(f'Máximo:   ${maximo:,}')
print(f'Mínimo:   ${minimo:,}')
````

## Argumentos con valor por defecto

Puedes definir valores por defecto para los parámetros. Si el usuario no los pasa, se usa el valor por defecto.

````python
def generar_reporte(empresa, moneda='USD', mostrar_detalle=True):
    """Genera encabezado de reporte con configuración opcional."""
    print(f'Empresa: {empresa} | Moneda: {moneda}')
    if mostrar_detalle:
        print('(Modo detallado activado)')

# Sin argumentos opcionales (usa los valores por defecto)
generar_reporte('TechCorp')
print()

# Con argumentos opcionales
generar_reporte('GlobalBank', moneda='EUR', mostrar_detalle=False)
````

## Variables locales vs globales

- **Variable local**: existe solo dentro de la función
- **Variable global**: existe en todo el programa

````python
# Variable global
tasa_iva_global = 0.15

def calcular_impuesto(precio):
    # Variable local (solo existe dentro de esta función)
    impuesto = precio * tasa_iva_global  # puede usar la variable global
    return impuesto

print(f'Impuesto sobre $200: ${calcular_impuesto(200):.2f}')

# 'impuesto' no existe fuera de la función
# print(impuesto)  # ← esto daría NameError
````

## Funciones Lambda — funciones anónimas de una línea

ChatGPT Image 13 may 2026, 03_43_39 p.m..png

````python
# Lambda: función simple en una línea
# Sintaxis: lambda parametros: expresion

calcular_iva = lambda precio: precio * 0.15
print(f'IVA de $500: ${calcular_iva(500):.2f}')

# Muy útil para ordenar listas por criterio personalizado
empleados = [
    {'nombre': 'Ana', 'salario': 4000},
    {'nombre': 'Luis', 'salario': 3200},
    {'nombre': 'María', 'salario': 5100}
]

# Ordenar por salario usando lambda
empleados_ordenados = sorted(empleados, key=lambda e: e['salario'], reverse=True)
for emp in empleados_ordenados:
    print(f"{emp['nombre']:8}: ${emp['salario']:,}")
````

## Funciones predefinidas importantes

Python tiene muchas funciones ya incluidas (built-in):

````python
datos = [15, 3, 8, 22, 7, 1, 19]

print('sum():', sum(datos))           # suma
print('min():', min(datos))           # mínimo
print('max():', max(datos))           # máximo
print('len():', len(datos))           # longitud
print('sorted():', sorted(datos))    # ordenado (nuevo objeto)
print('round():', round(3.14159, 2)) # redondear
print('abs():', abs(-42))            # valor absoluto
````

## ✅ Resumen del Módulo 9

- Las **funciones** se definen con `def nombre(params):`
- Usan `return` para devolver valores
- Pueden retornar **múltiples valores** (como tupla)
- **Parámetros con defecto**: `def f(x, tasa=0.15)`
- **Variables locales** existen solo dentro de la función
- **Lambda**: `lambda x: x * 2` para funciones simples de una línea
- **Docstrings** `"""..."""`: documenta qué hace la función
