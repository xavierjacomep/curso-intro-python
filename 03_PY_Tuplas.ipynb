## ¿Qué son las Tuplas?

Una **tupla** es una colección ordenada e **inmutable** de elementos. "Inmutable" significa que una vez creada, **no puedes modificar, agregar ni eliminar** sus elementos.

Las tuplas se crean con paréntesis `()` y los elementos se separan con comas.

**Analogía:** Una tupla es como un contrato firmado: una vez sellado, no puedes cambiar lo que dice. Es útil cuando los datos no deben modificarse (coordenadas, configuraciones, registros históricos).

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesType.png" width="750" align="center">

````python
# Crear una tupla con diferentes tipos de datos
# (string, entero, decimal)

producto = ('Laptop', 899, 4.5)
print(producto)
print('Tipo:', type(producto))
````

````python
# Ejemplo con datos de un empleado
empleado = ('Ana López', 'E-1042', 'Contabilidad', 3200.50)
print(empleado)
````

````python
# Las tuplas pueden contener cualquier combinación de tipos
mixta = ('texto', 42, 3.14, True, None)
print(mixta)
````

<a id="indexacion"></a>
## Indexación

Al igual que los strings, las tuplas son secuencias ordenadas y se accede a sus elementos con índices.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesIndex.gif" width="750" align="center">

````python
# Acceder a elementos por índice positivo
producto = ('Laptop', 899, 4.5, 'Electrónica')

print('Nombre:     ', producto[0])  # índice 0 → primer elemento
print('Precio:     ', producto[1])  # índice 1
print('Calificación:', producto[2]) # índice 2
print('Categoría:  ', producto[3])  # índice 3
````

````python
# También podemos ver el tipo de cada elemento
print(type(producto[0]))  # str
print(type(producto[1]))  # int
print(type(producto[2]))  # float
````

````python
# Indexación negativa — desde el final
print('Último elemento:     ', producto[-1])   # Electrónica
print('Penúltimo elemento:  ', producto[-2])   # 4.5
print('Primero (desde atrás):', producto[-4])   # Laptop
````

### Slicing de Tuplas

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/TuplesSlice.gif" width="750" align="center">

````python
# Slicing de tuplas
ventas_trimestre = ('Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio')

# Primer trimestre
print('Q1:', ventas_trimestre[0:3])

# Segundo trimestre
print('Q2:', ventas_trimestre[3:6])
````

<a id="operaciones"></a>
## Operaciones con Tuplas

````python
# Concatenar tuplas con +
datos_basicos   = ('Juan García', 'V-12345678')
datos_contacto  = ('juan@email.com', '0412-5551234')

datos_completos = datos_basicos + datos_contacto
print(datos_completos)
````

````python
# Longitud de una tupla
meses = ('Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic')
print('Número de meses:', len(meses))
````

````python
# Verificar si un elemento está en la tupla
categorias = ('Electrónica', 'Ropa', 'Alimentos', 'Muebles')

print('¿Está Ropa?:    ', 'Ropa' in categorias)
print('¿Está Juguetes?:', 'Juguetes' in categorias)
````

````python
# Ordenar elementos de una tupla
# (sorted() devuelve una lista, no una tupla)
precios = (450, 120, 890, 33, 750)
print('Precios ordenados:', sorted(precios))
print('Precio mínimo:', min(precios))
print('Precio máximo:', max(precios))
print('Precio promedio:', sum(precios) / len(precios))
````

````python
# Desempaquetar una tupla — asignar cada elemento a una variable
# Esto es muy útil en análisis de datos

registro = ('Ana Martínez', 'Ventas', 4500.75)
nombre, departamento, salario = registro  # desempaquetar

print(f'Nombre:      {nombre}')
print(f'Departamento:{departamento}')
print(f'Salario:     ${salario:,.2f}')
````

### Las tuplas son inmutables — no se pueden modificar

````python
# Intentar modificar una tupla genera un TypeError
producto = ('Laptop', 899)

# Descomenta la siguiente línea para ver el error:
# producto[1] = 950  # TypeError: 'tuple' object does not support item assignment

print('Las tuplas son inmutables — no puedes cambiar sus valores.')
print('Si necesitas modificar datos, usa una Lista.')
````

<a id="uso"></a>
## ¿Cuándo usar Tuplas?

Usa una **tupla** cuando:
- Los datos **no deben cambiar** (registros históricos, coordenadas, constantes)
- Quieres que sea evidente que esa colección es fija e invariable
- Necesitas usar la colección como **clave de un diccionario** (los diccionarios requieren claves inmutables)

Usa una **lista** cuando necesitas modificar, agregar o eliminar elementos (lo veremos en el siguiente módulo).

````python
# Ejemplo práctico: coordenadas geográficas (no cambian)
quito    = (-0.2295, -78.5243)  # (latitud, longitud)
bogota   = (4.7110,  -74.0721)
lima     = (-12.0464, -77.0428)

print(f'Quito:  lat={quito[0]}, lon={quito[1]}')
print(f'Bogotá: lat={bogota[0]}, lon={bogota[1]}')
print(f'Lima:   lat={lima[0]}, lon={lima[1]}')
````

````python
# Ejemplo: registros de transacciones (no se deben modificar)
transaccion = ('2024-01-15', 'FAC-0001', 'Cliente A', 1500.00, 'Pagado')

fecha, factura, cliente, monto, estado = transaccion
print(f'Factura {factura} | {fecha} | {cliente} | ${monto:.2f} | {estado}')
````

## ✅ Resumen del Módulo 3

- Una **tupla** es una colección ordenada e **inmutable** `(elem1, elem2, ...)`
- Se accede con índices `[i]` y se pueden extraer porciones con slicing `[i:j]`
- Soportan operaciones: `len()`, `min()`, `max()`, `sum()`, `sorted()`, `in`
- Se pueden **desempaquetar** asignando cada elemento a una variable
- Úsalas para datos que no deben modificarse
