## 1. 📄 Escritura de Archivos — Modo `'w'`

### 🏢 Analogía empresarial

> Imagina que tienes una hoja de papel en blanco y un bolígrafo. El **modo de escritura `'w'`** en Python es como tomar esa hoja, escribir lo que quieras, y guardarla con un nombre. Si ya existía una hoja con ese nombre, la **reemplazas completamente** con la nueva.

Para escribir un archivo, usamos la función `open()` con el argumento de modo establecido en `'w'` (**write** = escribir).

### Sintaxis básica:
```python
with open('nombre_archivo.txt', 'w', encoding='utf-8') as archivo:
    archivo.write("El texto que quieres guardar")
```

| Parámetro | Descripción |
|-----------|-------------|
| `'nombre_archivo.txt'` | Nombre (y ruta) del archivo a crear o reemplazar |
| `'w'` | Modo escritura (**w**rite) — crea el archivo si no existe |
| `encoding='utf-8'` | Codificación para soportar tildes y caracteres especiales |
| `.write()` | Método que escribe texto en el archivo |

````python
# Escribir una línea en un archivo
# Creamos el archivo 'reporte.txt' con una sola línea de texto

ruta_archivo = '/tmp/reporte.txt'

with open(ruta_archivo, 'w', encoding='utf-8') as archivo:
    archivo.write("Reporte de Ventas - Empresa XYZ")

print("✅ Archivo creado exitosamente.")
````

````python
# Leemos el archivo para verificar que se escribió correctamente

with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
    contenido = archivo.read()
    print("Contenido del archivo:")
    print(contenido)
````

## 2. 📄 Escribir Múltiples Líneas

Para escribir varias líneas, usamos **múltiples llamadas** a `.write()`, o incluimos el carácter de salto de línea `\n` al final de cada texto.

### ¿Qué es `\n`?

> El carácter `\n` es una **nueva línea** (como presionar Enter). Es invisible en el archivo, pero le dice al sistema que el siguiente texto debe comenzar en una línea diferente.

````python
# Escribir múltiples líneas usando \n (salto de línea)

with open(ruta_archivo, 'w', encoding='utf-8') as archivo:
    archivo.write("=== REPORTE DE VENTAS MENSUALES ===\n")
    archivo.write("Empresa: Distribuidora Andina S.A.\n")
    archivo.write("Mes: Octubre 2024\n")
    archivo.write("Región: Sierra Centro\n")

print("✅ Archivo con múltiples líneas creado.")
````

````python
# Verificamos el contenido

with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
    print(archivo.read())
````

## 3. 📋 Escribir una Lista de Líneas en un Archivo

En la práctica empresarial, los datos suelen estar almacenados en **listas**. Podemos recorrer una lista con un bucle `for` y escribir cada elemento en el archivo.

### 🏢 Caso empresarial:
Tienes una lista de transacciones del día y quieres guardarlas en un archivo de registro.

````python
# Lista de transacciones del día
transacciones = [
    "001 | Venta | Laptop Dell | $850.00\n",
    "002 | Venta | Monitor LG 24' | $320.00\n",
    "003 | Devolución | Teclado USB | -$45.00\n",
    "004 | Venta | Silla Ergonómica | $210.00\n",
    "005 | Venta | Proyector Epson | $620.00\n",
]

print("Lista de transacciones en memoria:")
for t in transacciones:
    print(t, end='')
````

````python
# Escribir la lista completa en el archivo
ruta_transacciones = '/tmp/transacciones_dia.txt'

with open(ruta_transacciones, 'w', encoding='utf-8') as archivo:
    archivo.write("=== REGISTRO DE TRANSACCIONES - 15/10/2024 ===\n")
    archivo.write("ID  | Tipo       | Producto              | Monto\n")
    archivo.write("-" * 55 + "\n")
    for transaccion in transacciones:
        archivo.write(transaccion)

print("✅ Registro de transacciones guardado.")
````

````python
# Verificar el archivo generado
with open(ruta_transacciones, 'r', encoding='utf-8') as archivo:
    print(archivo.read())
````

## 4. ⚠️ Advertencia: El Modo `'w'` Sobreescribe Todo

<div class="alert alert-block alert-danger">
<b>⚠️ PELIGRO:</b> El modo <code>'w'</code> <strong>borra todo el contenido existente</strong> del archivo antes de escribir. Si usas <code>'w'</code> sobre un archivo que ya tiene datos, esos datos se <strong>pierden para siempre</strong>.
</div>

### Demostración del riesgo:

````python
# DEMOSTRACIÓN: El modo 'w' sobreescribe el contenido anterior

ruta_demo = '/tmp/demo_sobreescritura.txt'

# Primero, creamos un archivo con información importante
with open(ruta_demo, 'w', encoding='utf-8') as f:
    f.write("Datos importantes de clientes:\n")
    f.write("Cliente 1: Empresa ABC - $15,000\n")
    f.write("Cliente 2: Corporación XYZ - $32,500\n")
    f.write("Cliente 3: Importadora Norte - $8,200\n")

print("📄 Archivo original:")
with open(ruta_demo, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# ⚠️ Usamos 'w' de nuevo — ¡esto BORRA todo lo anterior!

with open(ruta_demo, 'w', encoding='utf-8') as f:
    f.write("Texto de reemplazo\n")

print("📄 Archivo DESPUÉS de usar 'w' de nuevo:")
with open(ruta_demo, 'r', encoding='utf-8') as f:
    print(f.read())

print("\n⚠️  ¡Los datos de clientes anteriores se perdieron permanentemente!")
````

## 5. ➕ Agregar Contenido sin Borrar — Modo `'a'` (Append)

### 🏢 Analogía empresarial:

> El modo **`'a'`** (de **append** = añadir) es como un **libro de registro contable**: cada día agregas nuevas entradas al final, sin borrar los registros anteriores. Es perfecto para logs, bitácoras y reportes acumulativos.

| Modo | ¿Qué hace? | ¿Borra datos existentes? |
|------|-----------|-------------------------|
| `'w'` | Escribe desde cero | **SÍ** — peligroso si hay datos previos |
| `'a'` | Agrega al final | **NO** — los datos previos se conservan |

````python
# Creamos un archivo de log de ventas del mes
ruta_log = '/tmp/log_ventas.txt'

# Semana 1 - creamos el archivo por primera vez
with open(ruta_log, 'w', encoding='utf-8') as f:
    f.write("=== LOG DE VENTAS - OCTUBRE 2024 ===\n\n")
    f.write("[Semana 1]\n")
    f.write("Lunes 07/10: $4,200\n")
    f.write("Martes 08/10: $3,850\n")
    f.write("Miércoles 09/10: $5,100\n")

print("✅ Semana 1 registrada.")
with open(ruta_log, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# Semana 2 - AGREGAMOS al archivo existente con modo 'a'

with open(ruta_log, 'a', encoding='utf-8') as f:
    f.write("\n[Semana 2]\n")
    f.write("Lunes 14/10: $4,700\n")
    f.write("Martes 15/10: $6,200\n")
    f.write("Miércoles 16/10: $3,900\n")

print("✅ Semana 2 agregada. Contenido completo del log:")
print("-" * 40)
with open(ruta_log, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# Semana 3 - seguimos agregando

with open(ruta_log, 'a', encoding='utf-8') as f:
    f.write("\n[Semana 3]\n")
    f.write("Lunes 21/10: $5,500\n")
    f.write("Martes 22/10: $4,100\n")
    f.write("Miércoles 23/10: $7,300\n")
    f.write("\n--- FIN DEL REGISTRO ---\n")

print("✅ Log completo con 3 semanas:")
print("=" * 40)
with open(ruta_log, 'r', encoding='utf-8') as f:
    print(f.read())
````

## 6. 🔧 Modos Avanzados: `r+`, `w+`, `a+`

Python permite abrir archivos en modos que combinan lectura y escritura simultáneamente. Esto es útil cuando necesitas **leer y modificar** un archivo en la misma operación.

### Tabla completa de modos de apertura:

| Modo | Nombre | Lee | Escribe | ¿Borra contenido? | Posición inicial |
|------|--------|-----|---------|-------------------|------------------|
| `'r'` | Solo lectura | ✅ | ❌ | No | Inicio |
| `'w'` | Solo escritura | ❌ | ✅ | **Sí** | Inicio (vacío) |
| `'a'` | Solo agregar | ❌ | ✅ | No | **Final** |
| `'r+'` | Lectura y escritura | ✅ | ✅ | No | Inicio |
| `'w+'` | Escritura y lectura | ✅ | ✅ | **Sí** | Inicio (vacío) |
| `'a+'` | Agregar y lectura | ✅ | ✅ | No | **Final** |

### Concepto clave: el puntero de posición

> Piensa en el archivo como una **cinta de papel** y en el puntero como el **cursor** de tu editor de texto. Cuando lees o escribes, el cursor avanza. Para leer desde el inicio debes "mover el cursor" con el método `.seek(0)`.

````python
# Preparamos un archivo de ejemplo
ruta_avanzado = '/tmp/registro_avanzado.txt'

with open(ruta_avanzado, 'w', encoding='utf-8') as f:
    f.write("Registro inicial\n")
    f.write("Datos de la empresa\n")
    f.write("Fin del registro\n")

print("✅ Archivo de prueba creado.")
````

````python
# Modo 'a+': agregar texto Y poder leer después

with open(ruta_avanzado, 'a+', encoding='utf-8') as f:
    # Agregamos una línea nueva al final
    f.write("Nueva entrada agregada con a+\n")

    # El puntero quedó al FINAL después de escribir
    print(f"Posición del puntero después de escribir: {f.tell()}")

    # Para leer desde el inicio, movemos el puntero con seek(0)
    f.seek(0)
    print(f"Posición del puntero después de seek(0): {f.tell()}")

    # Ahora sí podemos leer todo el archivo
    contenido = f.read()

    if contenido:
        print("\n📄 Contenido completo del archivo:")
        print(contenido)
    else:
        print("(El archivo está vacío o el puntero está al final)")
````

````python
# Modo 'r+': leer y escribir sin borrar datos
# Útil para MODIFICAR partes específicas del archivo

ruta_rplus = '/tmp/datos_empresa.txt'

# Creamos el archivo
with open(ruta_rplus, 'w', encoding='utf-8') as f:
    f.write("Empresa: [PENDIENTE]\n")
    f.write("RUC: 1790000001001\n")
    f.write("Ciudad: Quito\n")

print("📄 Archivo antes de modificar:")
with open(ruta_rplus, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# Con r+ escribimos desde el inicio (reemplaza caracteres, no líneas)
with open(ruta_rplus, 'r+', encoding='utf-8') as f:
    f.seek(0)                          # Ir al inicio
    f.write("Empresa: TechCorp S.A.  ")  # Mismo número de caracteres para no desalinear

print("📄 Archivo después de modificar con r+:")
with open(ruta_rplus, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# Uso de .truncate() para cortar el archivo en la posición del puntero
# Útil para reescribir el inicio y eliminar el resto

ruta_truncate = '/tmp/reporte_truncado.txt'

with open(ruta_truncate, 'w', encoding='utf-8') as f:
    f.write("Línea 1: Ventas enero\n")
    f.write("Línea 2: Ventas febrero\n")
    f.write("Línea 3: Ventas marzo\n")
    f.write("Línea 4: BORRAR ESTE CONTENIDO\n")
    f.write("Línea 5: BORRAR ESTE TAMBIÉN\n")

# Reescribimos solo las primeras líneas y truncamos el resto
with open(ruta_truncate, 'r+', encoding='utf-8') as f:
    f.seek(0)
    f.write("Línea 1: Ventas enero    - $45,200\n")
    f.write("Línea 2: Ventas febrero  - $38,900\n")
    f.write("Línea 3: Ventas marzo    - $52,100\n")
    f.truncate()  # Elimina todo lo que viene después de la posición actual

print("📄 Archivo después de truncate:")
with open(ruta_truncate, 'r', encoding='utf-8') as f:
    print(f.read())
````

## 7. 📂 Copiar Archivos

Una operación muy común es **copiar** el contenido de un archivo a otro. En Python esto se hace abriendo dos archivos simultáneamente: uno para leer y otro para escribir.

### 🏢 Caso de uso:
- Crear respaldos (*backups*) de reportes importantes
- Hacer copias de bases de datos antes de modificarlas
- Duplicar plantillas de documentos

````python
# Creamos el archivo fuente (el original)
ruta_original = '/tmp/reporte_ventas_original.txt'
ruta_copia = '/tmp/reporte_ventas_BACKUP.txt'

with open(ruta_original, 'w', encoding='utf-8') as f:
    f.write("=== REPORTE DE VENTAS - Q3 2024 ===\n")
    f.write("Julio:      $125,400\n")
    f.write("Agosto:     $138,200\n")
    f.write("Septiembre: $142,700\n")
    f.write("\nTOTAL Q3:  $406,300\n")

print("✅ Archivo original creado.")
print("📄 Contenido original:")
with open(ruta_original, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# Copiar el archivo a un backup
# Abrimos ambos archivos al mismo tiempo: uno para leer, otro para escribir

with open(ruta_original, 'r', encoding='utf-8') as archivo_fuente:
    with open(ruta_copia, 'w', encoding='utf-8') as archivo_destino:
        for linea in archivo_fuente:
            archivo_destino.write(linea)

print("✅ Backup creado exitosamente.")
print(f"   Original: {ruta_original}")
print(f"   Copia:    {ruta_copia}")
````

````python
# Verificar que el backup es idéntico al original

print("📄 Contenido del BACKUP:")
print("=" * 35)
with open(ruta_copia, 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# Alternativa más eficiente: usar el módulo shutil
# shutil (shell utilities) tiene funciones especializadas para manejo de archivos

import shutil

ruta_copia2 = '/tmp/reporte_ventas_BACKUP2.txt'
shutil.copy(ruta_original, ruta_copia2)

print("✅ Copia realizada con shutil.copy()")

with open(ruta_copia2, 'r', encoding='utf-8') as f:
    print(f.read())
````

## 🔁 Aplicación Práctica: Generador de Reportes Automático

Veamos un caso integrador: una función que genera automáticamente un reporte de ventas y lo guarda en un archivo.

````python
def generar_reporte_ventas(datos_ventas, ruta_salida):
    """
    Genera un reporte de ventas y lo guarda en un archivo .txt

    Parámetros:
    - datos_ventas: lista de diccionarios con 'producto', 'cantidad' y 'precio_unit'
    - ruta_salida: ruta del archivo donde se guardará el reporte
    """
    total_general = 0

    with open(ruta_salida, 'w', encoding='utf-8') as f:
        # Encabezado del reporte
        f.write("=" * 55 + "\n")
        f.write("        REPORTE DE VENTAS - GENERADO AUTOMÁTICAMENTE\n")
        f.write("=" * 55 + "\n")
        f.write(f"{'PRODUCTO':<25} {'CANT':>5} {'P.UNIT':>10} {'TOTAL':>12}\n")
        f.write("-" * 55 + "\n")

        for item in datos_ventas:
            subtotal = item['cantidad'] * item['precio_unit']
            total_general += subtotal
            f.write(f"{item['producto']:<25} {item['cantidad']:>5} "
                    f"${item['precio_unit']:>9,.2f} ${subtotal:>11,.2f}\n")

        # Totales
        f.write("=" * 55 + "\n")
        f.write(f"{'TOTAL GENERAL':>41} ${total_general:>11,.2f}\n")
        f.write("=" * 55 + "\n")

    return total_general

# Datos de ejemplo
ventas_octubre = [
    {'producto': 'Laptop Dell Inspiron',    'cantidad': 12, 'precio_unit': 850.00},
    {'producto': 'Monitor LG 24"',          'cantidad': 25, 'precio_unit': 320.00},
    {'producto': 'Teclado Mecánico',        'cantidad': 40, 'precio_unit': 95.00},
    {'producto': 'Mouse Inalámbrico',       'cantidad': 55, 'precio_unit': 35.00},
    {'producto': 'Silla Ergonómica',        'cantidad': 8,  'precio_unit': 210.00},
    {'producto': 'Proyector Epson',         'cantidad': 3,  'precio_unit': 620.00},
    {'producto': 'Auriculares Sony',        'cantidad': 20, 'precio_unit': 125.00},
]

ruta_reporte = '/tmp/reporte_octubre_2024.txt'
total = generar_reporte_ventas(ventas_octubre, ruta_reporte)

print(f"✅ Reporte generado en: {ruta_reporte}")
print(f"💰 Total del período: ${total:,.2f}")
print()

# Mostrar el reporte generado
with open(ruta_reporte, 'r', encoding='utf-8') as f:
    print(f.read())
````

| `open('archivo', 'a+')` | Agrega y puede leer | Bitácora con consultas |
| `.write(texto)` | Escribe texto en el archivo | Guardar una línea de datos |
| `.seek(0)` | Mueve el puntero al inicio | Leer desde el principio tras escribir |
| `.tell()` | Informa la posición del puntero | Saber dónde está el cursor |
| `.truncate()` | Corta el archivo en la posición actual | Eliminar contenido obsoleto |
| `shutil.copy()` | Copia un archivo completo | Crear backups automáticos |
