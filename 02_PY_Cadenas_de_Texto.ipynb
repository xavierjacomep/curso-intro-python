<a id="indexacion"></a>
## Indexación — acceder a caracteres individuales

Un string es una secuencia **ordenada** de caracteres. Cada carácter tiene una posición llamada **índice**.

⚠️ **Importante:** En Python los índices comienzan en `0`, no en `1`.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/xwKqWxQWBL47h718d3BLzw/IMG1.png" width="600" align="center">

````python
# Definimos un string y accedemos a sus caracteres por índice
nombre = 'Economía'

print('Carácter en índice 0:', nombre[0])  # E
print('Carácter en índice 1:', nombre[1])  # c
print('Carácter en índice 4:', nombre[4])  # m
````

````python
# La función len() devuelve la longitud (número de caracteres) del string
nombre = 'Corporación'
print('Longitud:', len(nombre))
````

<a id="negativa"></a>
### Indexación Negativa

Python también permite usar índices negativos para contar desde el final:

- `-1` → último carácter
- `-2` → penúltimo carácter
- etc.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VAdKOVUWpsM7hC7CjWmEdQ/IMG2.png" width="600" align="center">

````python
nombre = 'Empresa'

print('Último carácter (índice -1):   ', nombre[-1])   # a
print('Penúltimo (índice -2):         ', nombre[-2])   # s
print('Primer carácter (índice -7):   ', nombre[-7])   # E  (len = 7)
````

<a id="slicing"></a>
### Slicing (rebanado) — extraer subcadenas

El **slicing** permite extraer una porción del string usando la notación `string[inicio:fin]`.

- El índice `inicio` está **incluido**
- El índice `fin` está **excluido** (no se incluye)

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/Ph9xvvIvaf-krPI-qoaZ2Q/IMG3.png" width="600" align="center">

````python
# Ejemplo de slicing
texto = 'Análisis de Datos'

# Extraer los primeros 8 caracteres (índices 0 al 7)
print(texto[0:8])   # → 'Análisis'

# Extraer 'Datos' (últimas 5 letras)
print(texto[12:17]) # → 'Datos'
````

````python
# Si omites el inicio, Python empieza desde 0
# Si omites el fin, Python va hasta el final

codigo = 'PROD-2024-001'
print('Categoría:', codigo[:4])    # → 'PROD'
print('Año:      ', codigo[5:9])   # → '2024'
print('Número:   ', codigo[10:])   # → '001'
````

<a id="stride"></a>
### Stride (paso) — seleccionar con saltos

La notación `string[inicio:fin:paso]` permite seleccionar caracteres con un intervalo específico.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/SaORQ3bZLArgeix_9-jjRQ/IMG4.png" width="600" align="center">

````python
# Seleccionar cada 2 caracteres
texto = 'Python'

print('Cada 2 caracteres:', texto[::2])   # P, t, o → 'Pto'
print('Cada 2, del 0 al 4:', texto[0:4:2]) # P, t → 'Pt'
````

````python
# Truco útil: invertir un string con paso -1
mensaje = 'datos'
print('Invertido:', mensaje[::-1])
````

<a id="concatenar"></a>
### Concatenar Strings

**Concatenar** significa unir dos strings. Se hace con el operador `+`.

````python
# Concatenación con +
nombre  = 'Juan'
apellido = 'García'
nombre_completo = nombre + ' ' + apellido
print(nombre_completo)
````

````python
# Repetir un string con *
separador = '-' * 30
print(separador)
print('REPORTE DE VENTAS')
print(separador)
````

<a id="escape"></a>
## Secuencias de Escape

Las secuencias de escape son combinaciones especiales que representan caracteres que no se pueden escribir directamente:

| Secuencia | Significado |
|-----------|-------------|
| `\n` | Nueva línea |
| `\t` | Tabulación (tab) |
| `\\` | Barra invertida `\` |
| `\'` | Comilla simple |
| `\"` | Comilla doble |

````python
# \n genera un salto de línea dentro del string
print('Línea 1\nLínea 2\nLínea 3')
````

````python
# \t genera un espacio de tabulación
print('Producto\tPrecio\tCantidad')
print('Laptop  \t$800  \t5')
print('Monitor \t$300  \t10')
````

````python
# Usar comilla dentro de un string
print("La empresa se llama 'Innovatech'")
print('El gerente dijo: "reunión mañana"')
````

<a id="metodos"></a>
## Métodos de Strings

Los strings tienen **métodos** incorporados: funciones específicas que puedes aplicar sobre un string usando la notación `string.metodo()`.

| Método | Descripción | Ejemplo |
|--------|-------------|--------|
| `.upper()` | Convierte a mayúsculas | `'hola'.upper()` → `'HOLA'` |
| `.lower()` | Convierte a minúsculas | `'HOLA'.lower()` → `'hola'` |
| `.replace(a, b)` | Reemplaza texto | `'hola'.replace('o','0')` → `'h0la'` |
| `.find(texto)` | Busca y devuelve el índice | `'hola'.find('la')` → `2` |
| `.strip()` | Elimina espacios al inicio y fin | `' hola '.strip()` → `'hola'` |
| `.split(sep)` | Divide el string | `'a,b,c'.split(',')` → `['a','b','c']` |
| `.startswith(t)` | ¿Empieza con...? | `'PROD-001'.startswith('PROD')` → `True` |

````python
# Ejemplos de métodos
nombre_cliente = '  Juan García  '

print('Original:    ', repr(nombre_cliente))
print('Sin espacios:', nombre_cliente.strip())
print('Mayúsculas:  ', nombre_cliente.strip().upper())
print('Minúsculas:  ', nombre_cliente.strip().lower())
````

````python
# Reemplazar texto — útil para limpiar datos
precio_texto = '$1,500.00'
precio_limpio = precio_texto.replace('$', '').replace(',', '')
precio_numero = float(precio_limpio)

print('Texto original:', precio_texto)
print('Como número:   ', precio_numero)
````

````python
# Split: dividir un string en partes
# Muy útil para procesar datos separados por comas (CSV)

fila_datos = 'Juan,García,Ventas,5000,Quito'
campos = fila_datos.split(',')

print('Todos los campos:', campos)
print('Nombre:', campos[0])
print('Departamento:', campos[2])
print('Ciudad:', campos[4])
````

````python
# Find: buscar texto dentro de un string
# Devuelve el índice donde empieza, o -1 si no lo encuentra

codigo = 'FACTURA-2024-00345'
pos = codigo.find('2024')
print('"2024" está en la posición:', pos)
````

<a id="fstrings"></a>
## f-strings — Formato moderno de texto

Los **f-strings** (formatted string literals) son una forma elegante de insertar variables dentro de un string. Se escriben con `f` antes de las comillas y las variables van dentro de `{}`.

Son muy útiles para generar reportes y mensajes automáticos.

````python
# Sin f-string (concatenación tradicional — más engorroso)
producto = 'Laptop'
precio = 899.99
print('El precio del ' + producto + ' es $' + str(precio))
````

````python
# Con f-string (moderno y más legible)
producto = 'Laptop'
precio = 899.99
print(f'El precio del {producto} es ${precio}')
````

````python
# f-strings con formato numérico
ventas = 125430.5678
print(f'Ventas del mes: ${ventas:,.2f}')   # formato con comas y 2 decimales

porcentaje = 0.1523
print(f'Tasa de crecimiento: {porcentaje:.1%}')  # como porcentaje
````

````python
# Ejemplo de reporte automático con f-strings
empresa = 'TechCorp'
mes = 'Enero'
ingresos = 85000
gastos = 52000
utilidad = ingresos - gastos

reporte = f"""
=== REPORTE MENSUAL ===
Empresa:  {empresa}
Mes:      {mes}
Ingresos: ${ingresos:,}
Gastos:   ${gastos:,}
Utilidad: ${utilidad:,}
Margen:   {utilidad/ingresos:.1%}
======================
"""
print(reporte)
````
