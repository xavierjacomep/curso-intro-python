## ¿Por qué necesitamos condiciones?

En la vida real, los programas deben tomar **decisiones** según los datos:

- Si las ventas superan la meta → enviar bono
- Si el inventario es menor a 10 unidades → hacer pedido automático
- Si el cliente es VIP → aplicar descuento del 20%

Las **estructuras condicionales** permiten que el código ejecute distintas acciones según se cumpla o no una condición.

## Operadores de Comparación

Comparan dos valores y devuelven `True` o `False`:

| Operador | Significado | Ejemplo | Resultado |
|----------|-------------|---------|----------|
| `==` | Igual a | `5 == 5` | `True` |
| `!=` | Diferente de | `5 != 3` | `True` |
| `>` | Mayor que | `10 > 7` | `True` |
| `<` | Menor que | `3 < 1` | `False` |
| `>=` | Mayor o igual | `5 >= 5` | `True` |
| `<=` | Menor o igual | `4 <= 3` | `False` |

````python
# Ejemplos de operadores de comparación
ventas = 5000
meta   = 6000

print('¿Ventas == Meta?:', ventas == meta)
print('¿Ventas < Meta?: ', ventas < meta)
print('¿Ventas >= Meta?:', ventas >= meta)
````

````python
# También funciona con strings
estado = 'activo'
print('¿Es activo?:', estado == 'activo')
print('¿Es inactivo?:', estado == 'inactivo')
````

## Estructura `if` — decisión simple

````python
# Estructura básica if
# El código INDENTADO dentro del if solo se ejecuta si la condición es True

stock = 5

if stock < 10:
    print('⚠️ Stock bajo — se necesita reabastecimiento')
    print(f'Cantidad actual: {stock} unidades')
````

<div class="alert alert-block alert-warning">
⚠️ <b>La indentación es CRÍTICA en Python.</b> El código dentro de un bloque <code>if</code> debe estar indentado (4 espacios o 1 Tab). Si no está indentado, Python genera un error.
</div>

## Estructura `if-else` — dos caminos

````python
ventas   = 7500
meta     = 6000
bono_pct = 0.10

if ventas >= meta:
    bono = ventas * bono_pct
    print(f'✅ Meta alcanzada. Bono: ${bono:,.2f}')
else:
    diferencia = meta - ventas
    print(f'❌ Meta no alcanzada. Faltaron ${diferencia:,}')
````

## Estructura `if-elif-else` — múltiples condiciones

````python
# Clasificar clientes por monto de compra anual
compra_anual = 12000

if compra_anual >= 20000:
    categoria = 'Platino'
    descuento = 0.25
elif compra_anual >= 10000:
    categoria = 'Oro'
    descuento = 0.15
elif compra_anual >= 5000:
    categoria = 'Plata'
    descuento = 0.10
else:
    categoria = 'Estándar'
    descuento = 0.05

print(f'Cliente {categoria}: {descuento:.0%} de descuento')
````

## Operadores Lógicos: `and`, `or`, `not`

Permiten combinar múltiples condiciones:

| Operador | Descripción | Ejemplo |
|----------|-------------|--------|
| `and` | Ambas deben ser True | `x > 0 and x < 10` |
| `or` | Al menos una True | `x < 0 or x > 100` |
| `not` | Invierte el valor | `not (x == 5)` |

````python
# Ejemplo: aprobación de crédito
ingresos_mensuales = 3500
historial_bueno    = True
deuda_existente    = 500

# Se aprueba si tiene buenos ingresos Y buen historial Y poca deuda
if ingresos_mensuales >= 2000 and historial_bueno and deuda_existente < 1000:
    print('✅ Crédito APROBADO')
else:
    print('❌ Crédito RECHAZADO — no cumple todos los criterios')
````

````python
# Alerta: cliente fuera de rango aceptable
edad = 17
if edad < 18 or edad > 100:
    print('⚠️ Edad fuera del rango aceptable')
else:
    print('Edad válida')
````

````python
# not — negar una condición
usuario_autenticado = False

if not usuario_autenticado:
    print('Por favor, inicia sesión para continuar')
````

## Condicional en una línea (ternario)

Para casos simples, Python permite escribir un `if-else` en una sola línea:

````python
# Sintaxis: valor_si_true if condicion else valor_si_false
saldo = 1500
estado = 'positivo' if saldo > 0 else 'negativo'
print(f'Saldo {estado}: ${saldo}')

# Otro ejemplo
nota = 75
resultado = 'Aprobado' if nota >= 60 else 'Reprobado'
print(f'Nota {nota}: {resultado}')
````
