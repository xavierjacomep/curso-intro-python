<a id="comentarios"></a>
### Comentarios en Python

Los **comentarios** son anotaciones dentro del código que Python ignora al ejecutarse. Sirven para:

- Explicar qué hace una parte del código
- Dejar notas para ti mismo o para otros
- "Desactivar" temporalmente una línea de código sin borrarla

Para escribir un comentario, usa el símbolo `#` antes del texto. Todo lo que esté después del `#` en esa línea será ignorado por Python.

````python
# Este es un comentario: Python lo ignora completamente

print('¡Hola, Python!')  # Este comentario está al final de una línea de código

# print('Esta línea está comentada y NO se ejecutará')
````

Observa que la segunda instrucción `print` (la que está comentada) **no produce ninguna salida**. Eso es porque Python la ignora.

Escribir buenos comentarios es una práctica muy valorada en el mundo profesional: hace que tu código sea más legible y fácil de mantener.

<a id="errores"></a>
### Errores en Python

Todos cometemos errores al programar. Python está diseñado para darte **mensajes de error claros** que te ayudan a identificar y corregir el problema. Es muy importante leer estos mensajes con atención.

Veamos un ejemplo: si escribimos `frint` en lugar de `print`, Python lanzará un error:

````python
# Ejemplo de error: nombre de función mal escrito
# 'frint' no existe en Python → genera un NameError

frint('¡Hola, Python!')
````

El mensaje de error nos dice:
1. **Dónde** ocurrió el error (número de línea)
2. **Qué tipo** de error fue (`NameError` — nombre no reconocido)

Python no conoce la función `frint` porque no está definida ni es parte del lenguaje.

Otro error común es olvidar cerrar las comillas de un texto:

````python
# Ejemplo de SyntaxError: comillas sin cerrar
# Descomenta la siguiente línea para ver el error:

# print("¡Hola, Python!)
````

Este produce un `SyntaxError` (error de sintaxis): el código no sigue las reglas gramaticales de Python.

### ¿Python detecta errores antes de ejecutar?

Python es un lenguaje **interpretado**: ejecuta el código línea por línea. Si encuentra un error en la línea 3, ejecutará las líneas 1 y 2, pero se detendrá en la 3. No revisa todo el código de antemano.

````python
# Demostración del orden de ejecución
# Python ejecuta línea por línea y se detiene en el error

print('Esta línea SÍ se imprimirá')   # línea 1: OK
frint('Esta línea causará un error')   # línea 2: ERROR → se detiene aquí
print('Esta línea NO se imprimirá')   # línea 3: nunca llega aquí
````

## 2. Tipos de Datos en Python

Python trabaja con diferentes **tipos de datos**. Los más fundamentales son:

| Tipo | Nombre en Python | Ejemplos |
|------|-----------------|----------|
| Números enteros | `int` | -10, 0, 42, 1000 |
| Números decimales | `float` | 3.14, -0.5, 2.0 |
| Texto (cadenas de caracteres) | `str` | "Hola", 'Python', "123" |
| Verdadero/Falso | `bool` | `True`, `False` |

En economía y administración usarás constantemente estos tipos: precios (float), cantidades (int), nombres de productos (str), condiciones (bool).

<a id="enteros"></a>
### Enteros (`int`)

Los enteros son números sin parte decimal. Pueden ser positivos o negativos.

````python
# Ejemplo de número entero
# En Python, simplemente escribes el número

11
````

````python
# La función type() nos dice de qué tipo es un dato

type(11)
````

<a id="flotantes"></a>
### Flotantes (`float`)

Los flotantes son números con parte decimal. El separador decimal en Python es el punto `.` (no la coma).

````python
# Ejemplo de número flotante (decimal)
# Importante: en Python usamos PUNTO como separador decimal, no coma

2.14
````

````python
# Verificamos el tipo

type(2.14)
````

````python
# El texto (string) también es un tipo de dato
# Siempre va entre comillas simples o dobles

"¡Hola, Python!"
````

````python
type("¡Hola, Python!")
````

<a id="conversion"></a>
### Conversión entre tipos de datos

A veces necesitas convertir un tipo de dato en otro. Por ejemplo, convertir un número a texto o viceversa. Python tiene funciones para esto:

| Función | Convierte a |
|---------|-------------|
| `int()` | Entero |
| `float()` | Decimal |
| `str()` | Texto |
| `bool()` | Booleano |

````python
# Convertir un flotante a entero (se elimina la parte decimal)
# Ejemplo: el precio 19.99 convertido a entero → 19

precio_decimal = 19.99
precio_entero = int(precio_decimal)
print('Decimal:', precio_decimal)
print('Entero:', precio_entero)
````

````python
# Convertir un entero a flotante

cantidad = 5
print('Entero:', cantidad)
print('Como decimal:', float(cantidad))
````

````python
# Convertir un número a texto (string)
# Útil para concatenar números con texto

ventas = 1500
mensaje = 'Las ventas del mes fueron: ' + str(ventas)
print(mensaje)
````

````python
# Convertir texto a número (solo si el texto representa un número)

texto_numero = '42'
print('Tipo original:', type(texto_numero))
numero = int(texto_numero)
print('Tipo convertido:', type(numero))
print('Valor:', numero)
````

<a id="booleanos"></a>
### Tipo Booleano (`bool`)

Los booleanos solo tienen dos valores: `True` (Verdadero) o `False` (Falso). Son fundamentales para la lógica de programación — los usaremos mucho en condiciones y decisiones.

**Nota importante:** En Python, `True` y `False` se escriben con la primera letra en mayúscula.

````python
# Ejemplo de booleano

es_rentable = True
tiene_deudas = False

print('¿Es rentable?', es_rentable)
print('¿Tiene deudas?', tiene_deudas)
print('Tipo de dato:', type(es_rentable))
````

````python
# Los booleanos pueden convertirse a números:
# True equivale a 1, False equivale a 0

print(int(True))   # → 1
print(int(False))  # → 0
````

## 3. Expresiones y Variables

<a id="operaciones"></a>
### Operaciones matemáticas

Python puede usarse como una calculadora. Los operadores matemáticos principales son:

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|----------|
| `+` | Suma | `5 + 3` | `8` |
| `-` | Resta | `10 - 4` | `6` |
| `*` | Multiplicación | `3 * 7` | `21` |
| `/` | División | `15 / 4` | `3.75` |
| `//` | División entera | `15 // 4` | `3` |
| `%` | Módulo (resto) | `15 % 4` | `3` |
| `**` | Potencia | `2 ** 8` | `256` |

````python
# Operaciones aritméticas básicas

print('Suma:          ', 43 + 60 + 16 + 41)
print('Resta:         ', 50 - 60)
print('Multiplicación:', 5 * 5)
print('División:      ', 25 / 5)
print('División entera:', 25 // 6)  # Descarta los decimales
print('Módulo (resto): ', 25 % 6)   # El resto de 25 ÷ 6
print('Potencia:      ', 2 ** 10)   # 2 elevado a la 10
````

````python
# Ejemplo de aplicación empresarial:
# Calcular el total de ventas de una tienda

precio_unitario = 25.50    # precio por unidad en dólares
unidades_vendidas = 120    # unidades vendidas en el mes
descuento = 0.10           # 10% de descuento

total_bruto = precio_unitario * unidades_vendidas
total_neto  = total_bruto * (1 - descuento)

print('Total bruto:  $', total_bruto)
print('Total neto:   $', total_neto)
````

<div class="alert alert-block alert-warning">
⚠️ <b>Importante:</b> Python respeta el orden de las operaciones matemáticas (PEMDAS): primero los paréntesis, luego potencias, multiplicación/división y por último suma/resta. Usa paréntesis si necesitas controlar el orden.
</div>

````python
# Los paréntesis controlan el orden de evaluación

print('Sin paréntesis:', 2 + 3 * 4)    # → 14  (3*4=12, luego +2)
print('Con paréntesis:', (2 + 3) * 4)  # → 20  (2+3=5, luego *4)
````

<a id="variables"></a>
### Variables

Una **variable** es un nombre que le damos a un espacio en memoria donde se almacena un valor. Piensa en ella como una caja etiquetada: el nombre es la etiqueta, y el valor es lo que hay dentro.

Para asignar un valor a una variable usamos el signo `=` (no es igualdad matemática, es **asignación**).

**Reglas para nombrar variables:**
- Pueden contener letras, números y guión bajo `_`
- No pueden comenzar con un número
- No pueden contener espacios (usa `_` en su lugar)
- Son sensibles a mayúsculas (`precio` ≠ `Precio`)
- Evita usar palabras reservadas de Python (`if`, `for`, `print`, etc.)

````python
# Asignar valores a variables

mi_variable = 1        # entero
mi_variable
````

````python
# El valor de una variable puede cambiar (por eso se llama "variable")

mi_variable = 10
mi_variable
````

````python
# Ejemplo empresarial con variables
# Calcular la utilidad de un proyecto

ingresos    = 50000    # dólares
costos      = 30000    # dólares
impuestos   = 0.22     # 22% de impuesto sobre la utilidad

utilidad_bruta = ingresos - costos
impuesto_valor = utilidad_bruta * impuestos
utilidad_neta  = utilidad_bruta - impuesto_valor

print('=== Estado de Resultados Simplificado ===')
print('Ingresos:        $', ingresos)
print('Costos:          $', costos)
print('Utilidad bruta:  $', utilidad_bruta)
print('Impuestos (22%): $', impuesto_valor)
print('Utilidad neta:   $', utilidad_neta)
````
