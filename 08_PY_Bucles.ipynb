## ¿Por qué necesitamos bucles?

## La función `range()`

`range(n)` genera una secuencia de números del 0 al n-1.

````python
# range(5) genera: 0, 1, 2, 3, 4
print(list(range(5)))

# range(inicio, fin) genera desde inicio hasta fin-1
print(list(range(1, 6)))    # 1, 2, 3, 4, 5

# range(inicio, fin, paso)
print(list(range(0, 20, 5)))  # 0, 5, 10, 15
````

## Bucle `for`

El bucle `for` recorre (itera sobre) una secuencia: lista, string, range, etc.

````python
# Iterar sobre una lista de productos
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor']

for producto in productos:
    print(f'Verificando inventario de: {producto}')
````

````python
# Iterar con índice usando enumerate()
# enumerate() devuelve el índice y el valor

for indice, producto in enumerate(productos):
    print(f'{indice + 1}. {producto}')
````

````python
# Calcular el total de ventas iterando
ventas_diarias = [1200, 980, 1500, 2100, 750, 1800, 1350]
total = 0

for venta in ventas_diarias:
    total += venta   # equivale a: total = total + venta

print(f'Total de la semana: ${total:,}')
print(f'Promedio diario:    ${total/len(ventas_diarias):,.2f}')
````

````python
# Iterar sobre un diccionario
precios = {'Laptop': 899, 'Mouse': 25, 'Teclado': 75, 'Monitor': 350}

print('=== LISTA DE PRECIOS CON IVA (15%) ===')
for producto, precio in precios.items():
    precio_iva = precio * 1.15
    print(f'{producto:10}: ${precio:>5} → ${precio_iva:>8.2f} con IVA')
````

````python
# Iterar sobre un string carácter por carácter
codigo = 'FACTURA'
for letra in codigo:
    print(letra, end='-')  # end='-' evita el salto de línea
print()  # salto de línea al final
````

## Bucle `while`

El bucle `while` se ejecuta **mientras** una condición sea `True`. Útil cuando no sabes cuántas veces necesitas repetir.

````python
# Ejemplo: amortización de un préstamo
saldo = 10000    # deuda inicial
pago_mensual = 1500
tasa_mensual = 0.02  # 2% mensual
mes = 0

print('Mes | Interés | Pago    | Saldo')
print('-' * 45)

while saldo > 0:
    mes += 1
    interes = saldo * tasa_mensual
    pago_real = min(pago_mensual, saldo + interes)  # no pagar más de lo que se debe
    saldo = saldo + interes - pago_real
    saldo = max(0, saldo)  # que no quede negativo por redondeo
    print(f'{mes:>3} | ${interes:>7.2f} | ${pago_real:>7.2f} | ${saldo:>8.2f}')
    if mes > 24:  # protección: máximo 24 meses
        break

print(f'\nPréstamo pagado en {mes} meses')
````

## Control de bucles: `break` y `continue`

- **`break`**: detiene el bucle completamente
- **`continue`**: salta a la siguiente iteración (omite el resto del código en esta vuelta)

````python
# break — detenerse cuando se encuentra el primer producto sin stock
inventario = [
    ('Laptop', 10),
    ('Mouse', 25),
    ('Teclado', 0),    # sin stock
    ('Monitor', 8),
    ('Webcam', 0)      # sin stock
]

print('Revisando inventario...')
for producto, stock in inventario:
    if stock == 0:
        print(f'⚠️ Primer producto sin stock: {producto} — ALERTA ENVIADA')
        break  # detiene el bucle en el primero sin stock
    print(f'  ✅ {producto}: {stock} unidades')
````

````python
# continue — saltar transacciones nulas
transacciones = [500, 0, 1200, 0, 850, 320, 0, 950]

total_real = 0
for monto in transacciones:
    if monto == 0:
        continue  # saltar a la siguiente sin procesar el 0
    total_real += monto

print(f'Total (ignorando transacciones nulas): ${total_real:,}')
````

## ✅ Resumen del Módulo 8

- **`for`**: itera sobre una secuencia (lista, string, range, diccionario)
- **`range(inicio, fin, paso)`**: genera secuencias numéricas
