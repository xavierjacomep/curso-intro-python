## Crear Listas e Indexar

Una lista se crea con corchetes `[]` y los elementos se separan con comas.

ListsIndex_es.png

````python
# Crear una lista con diferentes tipos de datos
registro = ['Laptop', 899.99, 2024, True]
print(registro)
print('Tipo:', type(registro))
````

````python
# Acceso por índice positivo y negativo
print('Primer elemento (índice 0): ', registro[0])
print('Último elemento (índice -1):', registro[-1])
print('Longitud de la lista:       ', len(registro))
````

````python
# Lista de ventas mensuales
ventas = [5200, 6100, 4800, 7300, 6900, 8200]

# Slicing — obtener los primeros 3 meses
primer_trimestre = ventas[0:3]
print('Primer trimestre:', primer_trimestre)
````

<a id="contenido"></a>
## Contenido de las Listas — listas anidadas

Las listas pueden contener **cualquier tipo de dato**, incluyendo otras listas y tuplas (listas anidadas).

````python
# Lista anidada — como una tabla de datos
tabla_productos = [
    ['Laptop', 899, 'Electrónica'],
    ['Escritorio', 250, 'Muebles'],
    ['Cuaderno', 5, 'Papelería']
]

print(tabla_productos)
print()
# Acceder a un elemento de la lista anidada
print('Segundo producto:', tabla_productos[1])        # → ['Escritorio', 250, 'Muebles']
print('Precio del segundo:', tabla_productos[1][1])   # → 250
````

<a id="operaciones"></a>
## Operaciones: Agregar y Eliminar Elementos

Esta es la ventaja clave de las listas sobre las tuplas: podemos modificarlas.

### `append()` — agregar un elemento al final

````python
# append() agrega UN elemento al final de la lista
carrito = ['Laptop', 'Mouse']
print('Antes:', carrito)

carrito.append('Teclado')
print('Después:', carrito)
````

### `extend()` — agregar múltiples elementos

````python
# extend() agrega MÚLTIPLES elementos (los elementos de otra lista)
carrito = ['Laptop', 'Mouse']
nuevos = ['Monitor', 'Webcam', 'Cable HDMI']

carrito.extend(nuevos)
print('Después de extend:', carrito)
````

````python
# ⚠️ Diferencia importante: append vs extend con una lista
lista_a = ['a', 'b']
lista_b = ['a', 'b']

lista_a.append(['c', 'd'])   # agrega la lista COMO UN SOLO ELEMENTO
lista_b.extend(['c', 'd'])   # agrega los elementos INDIVIDUALMENTE

print('Con append:', lista_a)  # ['a', 'b', ['c', 'd']]
print('Con extend:', lista_b)  # ['a', 'b', 'c', 'd']
````

### Modificar elementos

````python
# Las listas SÍ permiten modificar elementos (a diferencia de las tuplas)
precios = [100, 200, 300]
print('Antes:', precios)

precios[1] = 250   # cambiar el segundo precio
print('Después:', precios)
````

### Eliminar elementos

````python
# del — eliminar por índice
inventario = ['Laptop', 'Mouse', 'Teclado', 'Monitor']
del inventario[1]  # elimina 'Mouse'
print('Después de del:', inventario)
````

````python
# remove() — eliminar por valor
inventario = ['Laptop', 'Mouse', 'Teclado', 'Monitor']
inventario.remove('Mouse')  # elimina la primera aparición de 'Mouse'
print('Después de remove:', inventario)
````

````python
# pop() — elimina y devuelve el último elemento (o el del índice especificado)
cola_pedidos = ['Pedido-001', 'Pedido-002', 'Pedido-003']
procesado = cola_pedidos.pop(0)  # elimina y retorna el primero

print('Pedido procesado:', procesado)
print('Cola restante:', cola_pedidos)
````

### Otros métodos útiles

````python
# sort() — ordenar la lista IN PLACE (modifica la lista original)
precios = [450, 120, 890, 33, 750]
precios.sort()
print('Ordenado ascendente:', precios)

precios.sort(reverse=True)
print('Ordenado descendente:', precios)
````

````python
# index() — encontrar la posición de un elemento
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor']
pos = productos.index('Teclado')
print('Teclado está en el índice:', pos)
````

````python
# count() — contar cuántas veces aparece un elemento
calificaciones = ['A', 'B', 'A', 'C', 'A', 'B']
print('Veces que aparece A:', calificaciones.count('A'))
````

````python
# Concatenar listas con +
ventas_q1 = [5000, 6000, 5500]
ventas_q2 = [7000, 6800, 7500]
ventas_semestre = ventas_q1 + ventas_q2

print('Semestre completo:', ventas_semestre)
print('Total semestre:   ', sum(ventas_semestre))
print('Promedio mensual: ', sum(ventas_semestre) / len(ventas_semestre))
````

<a id="clonar"></a>
## Copiar y Clonar Listas

⚠️ **Trampa común:** En Python, si copias una lista con `=`, **no creas una copia**, sino una **referencia** al mismo objeto. Cualquier cambio en una afectará a la otra.

````python
# ❌ MAL: asignación con = no crea una copia real
lista_original = [1, 2, 3]
lista_copia    = lista_original   # ambas apuntan al MISMO objeto

lista_copia.append(4)
print('Original:', lista_original)  # ← ¡también cambió!
print('"Copia": ', lista_copia)
````

````python
# ✅ BIEN: clonar con [:] crea una copia independiente
lista_original = [1, 2, 3]
lista_clon     = lista_original[:]  # slice vacío = copia completa

lista_clon.append(4)
print('Original:', lista_original)  # ← no cambió
print('Clon:    ', lista_clon)
````

````python
# También puedes usar .copy() — más explícito
original = [10, 20, 30]
copia = original.copy()
copia.append(40)

print('Original:', original)  # intacto
print('Copia:   ', copia)
````

<a id="comprension"></a>
## List Comprehensions — listas en una línea

Las **list comprehensions** son una forma elegante y compacta de crear listas. Son muy usadas en análisis de datos.

Sintaxis: `[expresión for elemento in iterable]`

````python
# Forma tradicional: crear lista de cuadrados
cuadrados = []
for n in range(1, 6):
    cuadrados.append(n ** 2)
print('Tradicional:', cuadrados)
````

````python
# Con list comprehension — mucho más conciso
cuadrados = [n ** 2 for n in range(1, 6)]
print('List comprehension:', cuadrados)
````

````python
# Con condición — solo incluir ciertos elementos
precios = [120, 45, 890, 230, 67, 540, 12]
precios_caros = [p for p in precios if p > 200]
print('Precios mayores a $200:', precios_caros)
````

````python
# Aplicación: calcular IVA del 15% para cada precio
precios_sin_iva = [100, 250, 75, 400, 180]
precios_con_iva = [round(p * 1.15, 2) for p in precios_sin_iva]

print('Sin IVA:', precios_sin_iva)
print('Con IVA:', precios_con_iva)
````

## ✅ Resumen del Módulo 4

- Una **lista** es una colección ordenada y **mutable** `[elem1, elem2, ...]`
- Indexación con `[i]`, slicing con `[i:j]`, igual que tuplas y strings
- Métodos clave: `append()`, `extend()`, `remove()`, `pop()`, `sort()`, `index()`, `count()`
- Para copiar: usa `[:]` o `.copy()`, nunca `=` directamente
- **List comprehensions** para crear listas de forma concisa
