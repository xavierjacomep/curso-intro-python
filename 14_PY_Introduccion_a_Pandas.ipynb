## 2. 🐼 ¿Qué es Pandas?

**Pandas** (Python Data Analysis Library) es la biblioteca más utilizada en el mundo para análisis de datos con Python. El nombre viene de **Pan**el **Da**ta (datos de panel), un término econométrico.

### La estructura principal: el DataFrame

> Un **DataFrame** es como una hoja de cálculo de Excel: tiene filas, columnas, encabezados, y puedes aplicar operaciones a sus datos. La diferencia es que con Python puedes automatizar todo.

```
EXCEL:                    PANDAS:
┌──────────┬──────────┐   ┌──────────┬──────────┐
│ Producto │ Precio   │   │ Producto │ Precio   │
├──────────┼──────────┤   ├──────────┼──────────┤
│ Laptop   │ $850     │   │ Laptop   │  850.00  │
│ Monitor  │ $320     │   │ Monitor  │  320.00  │
└──────────┴──────────┘   └──────────┴──────────┘
```

Por convención universal, Pandas se importa con el alias `pd`:

<a id="columnas"></a>
## 6. 📌 Acceder a Columnas

En Pandas, acceder a una columna es como acceder a un valor en un diccionario: usas el nombre de la columna como clave.

### Dos formas de acceder a columnas:

| Sintaxis | Resultado | Cuándo usar |
|----------|-----------|-------------|
| `df['Columna']` | **Serie** (1 dimensión) | Para operaciones, cálculos, filtros |
| `df[['Columna']]` | **DataFrame** (2 dimensiones) | Para mantener formato tabular |

````python
# Acceder a una columna como Serie (un corchete)
# Una Serie es como una lista con índice

productos = df['Product']
print("Tipo:", type(productos))
print()
print(productos)
````

````python
# Acceder a una columna como DataFrame (doble corchete)
# Retorna una tabla de una sola columna

cantidades = df[['Quantity']]
print("Tipo:", type(cantidades))
print()
cantidades
````

````python
# Acceder a MÚLTIPLES columnas (siempre doble corchete)
# Pasamos una lista de nombres de columnas

resumen = df[['Product', 'Category', 'Quantity', 'Total']]
resumen
````

````python
# Operaciones directas sobre columnas
# Las Series soportan operaciones matemáticas vectorizadas

print("💰 Resumen de la columna 'Total':")
print(f"   Suma total de ventas: ${df['Total'].sum():,.2f}")
print(f"   Venta promedio:       ${df['Total'].mean():,.2f}")
print(f"   Venta máxima:         ${df['Total'].max():,.2f}")
print(f"   Venta mínima:         ${df['Total'].min():,.2f}")
````

<a id="iloc-loc"></a>
## 7. 🎯 Acceder a Filas y Celdas con `iloc` y `loc`

Para acceder a elementos específicos del DataFrame necesitas indicar tanto la **fila** como la **columna**. Pandas ofrece dos métodos:

| Método | Selecciona por | Ejemplo |
|--------|---------------|--------|
| `df.iloc[fila, columna]` | **Posición numérica** (0, 1, 2...) | `df.iloc[0, 0]` → primera fila, primera columna |
| `df.loc[fila, 'Columna']` | **Nombre/etiqueta** | `df.loc[0, 'Product']` → fila 0, columna 'Product' |

### Analogía:
> `iloc` es como navegar por posición en un tablero de ajedrez: "columna 3, fila 5".  
> `loc` es como buscar en un mapa con nombre: "busca en calle 'Product', número 0".

````python
# Referencia visual del DataFrame
print("Vista del DataFrame (primeras 5 filas):")
df.head()
````

````python
# iloc: acceso por posición numérica
# Recordar: en Python los índices empiezan en 0

print("🔢 Acceso con iloc (por posición):")
print(f"  Fila 0, Columna 0 (primera celda):       {df.iloc[0, 0]}")
print(f"  Fila 1, Columna 0 (segunda fila, OrderID): {df.iloc[1, 0]}")
print(f"  Fila 0, Columna 2 (Category del pedido 1): {df.iloc[0, 2]}")
print(f"  Fila 1, Columna 2 (Category del pedido 2): {df.iloc[1, 2]}")
````

````python
# iloc: también puede seleccionar una fila completa
print("Fila completa del primer pedido:")
print(df.iloc[0])
````

````python
# loc: acceso por etiqueta (nombre de columna)
# La sintaxis es: df.loc[índice_fila, 'NombreColumna']

print("🏷️ Acceso con loc (por nombre):")
print(f"  Fila 0, columna 'Product':     {df.loc[0, 'Product']}")
print(f"  Fila 1, columna 'Product':     {df.loc[1, 'Product']}")
print(f"  Fila 1, columna 'CustomerCity': {df.loc[1, 'CustomerCity']}")
print(f"  Fila 1, columna 'Total':        ${df.loc[1, 'Total']:,.2f}")
````

````python
# Comparación visual: iloc vs loc para el mismo dato
# El pedido en la posición 4 (índice 4)

print("Mismo dato, dos formas:")
print(f"  df.iloc[4, 5] → {df.iloc[4, 5]}")
print(f"  df.loc[4, 'CustomerCity'] → {df.loc[4, 'CustomerCity']}")
print("  ✅ Mismo resultado, distinta sintaxis")
````

<a id="slicing"></a>
## 8. ✂️ Slicing — Selección de Rangos

Al igual que con las listas en Python, puedes seleccionar **rangos** de filas y columnas en un DataFrame.

### Diferencia importante entre `iloc` y `loc` en slicing:

| | `iloc` | `loc` |
|-|--------|-------|
| `[0:3]` | Filas 0, 1, 2 (excluye el 3) | Filas con índice 0, 1, 2, **3** (incluye el último) |

````python
# Slicing con iloc: seleccionar primeras 3 filas, primeras 3 columnas
print("Primeras 3 filas × primeras 3 columnas (iloc):")
df.iloc[0:3, 0:3]
````

````python
# Slicing con loc: seleccionar por nombre de columna
# loc es inclusivo en ambos extremos
print("Filas 0 a 3, columnas de OrderID a Category (loc):")
df.loc[0:3, 'OrderID':'Category']
````

````python
# Selección de filas específicas con lista de índices
print("Filas 0, 2 y 4 (tres pedidos específicos):")
df.iloc[[0, 2, 4]]
````

````python
# Seleccionar todas las filas de ciertas columnas
# El : solo significa "todas las filas"
print("Todas las filas, solo columnas Product y Total:")
df.loc[:, ['Product', 'Total']]
````

## 9. 🔎 Filtrar Datos con Condiciones

Una de las funcionalidades más poderosas de Pandas es poder **filtrar filas** basándose en condiciones. Es como la función FILTRO de Excel, pero mucho más flexible.

### Sintaxis:
```python
df[df['Columna'] condición valor]
```

### Operadores de comparación:

| Operador | Significado | Ejemplo |
|----------|-------------|--------|
| `==` | Igual a | `df['City'] == 'Quito'` |
| `!=` | Diferente de | `df['Category'] != 'Periféricos'` |
| `>` | Mayor que | `df['Total'] > 2000` |
| `>=` | Mayor o igual | `df['Quantity'] >= 10` |
| `<` | Menor que | `df['Price'] < 200` |
| `<=` | Menor o igual | `df['Price'] <= 500` |

### Operadores lógicos:
| Operador | Pandas | Ejemplo |
|----------|--------|--------|
| Y (AND) | `&` | `(df['Total'] > 1000) & (df['City'] == 'Quito')` |
| O (OR) | `\|` | `(df['Category'] == 'Audio') \| (df['Category'] == 'Móviles')` |
| NO (NOT) | `~` | `~(df['Category'] == 'Periféricos')` |

````python
# Filtro simple: ventas con Total mayor a $2,000
ventas_grandes = df[df['Total'] > 2000]
print(f"📊 Ventas superiores a $2,000: {len(ventas_grandes)} pedidos")
ventas_grandes
````

````python
# Filtro por texto: ventas en Quito
ventas_quito = df[df['CustomerCity'] == 'Quito']
print(f"🏙️ Ventas en Quito: {len(ventas_quito)} pedidos")
print(f"   Total vendido en Quito: ${ventas_quito['Total'].sum():,.2f}")
ventas_quito
````

````python
# Filtro por categoría
perifericos = df[df['Category'] == 'Periféricos']
print(f"🖱️ Productos en categoría 'Periféricos': {len(perifericos)}")
perifericos[['Product', 'Quantity', 'Price', 'Total']]
````

````python
# Filtro con dos condiciones (AND): ventas en Quito Y mayores a $1,000
# IMPORTANTE: cada condición debe ir entre paréntesis

quito_grandes = df[(df['CustomerCity'] == 'Quito') & (df['Total'] > 1000)]
print(f"🏙️💰 Ventas grandes en Quito: {len(quito_grandes)} pedidos")
quito_grandes[['Product', 'CustomerCity', 'Total']]
````

````python
# Filtro OR: ventas en Guayaquil O Cuenca
costa_sierra = df[(df['CustomerCity'] == 'Guayaquil') | (df['CustomerCity'] == 'Cuenca')]
print(f"🌊⛰️ Pedidos de Guayaquil o Cuenca: {len(costa_sierra)}")
costa_sierra[['Product', 'CustomerCity', 'Total']]
````

````python
# isin(): filtrar por múltiples valores (equivalente a OR múltiple)
ciudades_principales = df[df['CustomerCity'].isin(['Quito', 'Guayaquil', 'Cuenca'])]
print(f"🏙️ Pedidos de las 3 ciudades principales: {len(ciudades_principales)}")
ciudades_principales[['Product', 'CustomerCity', 'Total']]
````

````python
# str.contains(): buscar texto parcial en una columna
laptops = df[df['Product'].str.contains('Laptop')]
print(f"💻 Productos que contienen 'Laptop': {len(laptops)}")
laptops[['Product', 'Quantity', 'Price', 'Total']]
````

## 10. 📈 Estadísticas Descriptivas y Agrupaciones

Pandas facilita el cálculo de estadísticas y la agrupación de datos, algo fundamental en economía y análisis empresarial.

````python
# Estadísticas básicas por columna
print("📊 Resumen ejecutivo de ventas:")
print(f"  Total de pedidos:          {len(df)}")
print(f"  Ingresos totales:          ${df['Total'].sum():>12,.2f}")
print(f"  Promedio por pedido:       ${df['Total'].mean():>12,.2f}")
print(f"  Pedido más grande:         ${df['Total'].max():>12,.2f}")
print(f"  Pedido más pequeño:        ${df['Total'].min():>12,.2f}")
print(f"  Desviación estándar:       ${df['Total'].std():>12,.2f}")
print(f"  Unidades vendidas (total): {df['Quantity'].sum():>12,}")
````

````python
# groupby(): agrupar datos por una categoría
# Equivalente a la Tabla Dinámica de Excel

print("📊 Ventas totales por ciudad:")
ventas_ciudad = df.groupby('CustomerCity')['Total'].sum().sort_values(ascending=False)
print(ventas_ciudad.apply(lambda x: f"${x:,.2f}"))
````

````python
# Estadísticas múltiples por grupo
print("📊 Análisis por categoría de producto:")
analisis_categoria = df.groupby('Category').agg(
    num_pedidos=('OrderID', 'count'),
    unidades_totales=('Quantity', 'sum'),
    ingresos_totales=('Total', 'sum'),
    promedio_pedido=('Total', 'mean')
).round(2).sort_values('ingresos_totales', ascending=False)

analisis_categoria
````

````python
# value_counts(): contar valores únicos en una columna
print("🏙️ Número de pedidos por ciudad:")
print(df['CustomerCity'].value_counts())
````

````python
# Crear una nueva columna calculada
# Agregamos el precio con descuento del 10%

df['Price_con_descuento'] = df['Price'] * 0.90
df['Total_con_descuento'] = df['Price_con_descuento'] * df['Quantity']

print("💡 DataFrame con columnas calculadas:")
df[['Product', 'Price', 'Price_con_descuento', 'Total', 'Total_con_descuento']].head(5)
````

````python
# Ordenar el DataFrame por una columna
print("🏆 Top 5 productos por ingreso total:")
df.sort_values('Total', ascending=False)[['Product', 'Category', 'Quantity', 'Total']].head(5)
````

sobre_promedio = df[df['Total'] > promedio]
print(f"\n3️⃣ Promedio general: ${promedio:,.2f}")
print(f"   Pedidos sobre el promedio: {len(sobre_promedio)}")

# 4. Ingreso de Computación
total_computacion = df[df['Category'] == 'Computación']['Total'].sum()
print(f"\n4️⃣ Ingreso total 'Computación': ${total_computacion:,.2f}")

# 5. Exportar Guayaquil
df[df['CustomerCity'] == 'Guayaquil'].to_csv('/tmp/pedidos_guayaquil.csv', index=False)

## 📊 Resumen del Módulo

| Función/Método | Descripción | Ejemplo |
|----------------|-------------|--------|
| `pd.read_csv('archivo.csv')` | Carga un CSV | `df = pd.read_csv('ventas.csv')` |
| `pd.read_excel('archivo.xlsx')` | Carga un Excel | `df = pd.read_excel('informe.xlsx')` |
| `df.head(n)` | Primeras n filas | `df.head(10)` |
| `df.tail(n)` | Últimas n filas | `df.tail(5)` |
| `df.shape` | Dimensiones (filas, cols) | `(100, 7)` |
| `df.info()` | Info completa del df | tipos, nulos, memoria |
| `df.describe()` | Estadísticas descriptivas | media, std, min, max |
| `df['Columna']` | Seleccionar columna como Serie | `df['Total']` |
| `df[['Col1', 'Col2']]` | Seleccionar múltiples columnas | `df[['Product', 'Total']]` |
| `df.iloc[f, c]` | Celda por posición numérica | `df.iloc[0, 0]` |
| `df.loc[f, 'Col']` | Celda por nombre de columna | `df.loc[0, 'Product']` |
| `df[condición]` | Filtrar filas | `df[df['Total'] > 2000]` |
| `df.groupby('Col').sum()` | Agrupar y agregar | Tabla dinámica |
| `df.sort_values('Col')` | Ordenar por columna | ranking de ventas |
| `df.to_csv('salida.csv')` | Exportar a CSV | guardar resultados |
| `df.to_excel('salida.xlsx')` | Exportar a Excel | informe ejecutivo |
