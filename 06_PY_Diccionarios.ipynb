## ¿Qué es un Diccionario?

Un **diccionario** almacena pares **clave: valor**. A diferencia de las listas (que usan índices numéricos), los diccionarios usan **claves descriptivas** para acceder a los valores.

**Analogía:** Como un directorio telefónico: buscas por **nombre** (clave) y obtienes el **número** (valor). No necesitas saber en qué posición está.

Se crean con llaves `{}` y dos puntos `:` entre clave y valor.

````python
# Crear un diccionario de información de producto
producto = {
    'nombre': 'Laptop ProX',
    'precio': 1200.99,
    'stock': 45,
    'categoria': 'Electrónica',
    'activo': True
}
print(producto)
````

````python
# Acceder a valores por clave
print('Nombre:   ', producto['nombre'])
print('Precio:   ', producto['precio'])
print('Stock:    ', producto['stock'])
````

````python
# Método get() — más seguro que [] porque no da error si la clave no existe
print(producto.get('precio'))          # → 1200.99
print(producto.get('descuento'))       # → None (clave no existe, sin error)
print(producto.get('descuento', 0))    # → 0 (valor por defecto)
````

````python
# Ver todas las claves y valores
print('Claves:  ', list(producto.keys()))
print('Valores: ', list(producto.values()))
print('Pares:   ', list(producto.items()))
````

````python
# Agregar y modificar entradas
producto['descuento'] = 0.10     # agregar nueva clave
producto['stock'] = 40           # modificar valor existente

print('Con descuento y stock actualizado:')
print(producto)
````

````python
# Eliminar entradas
del producto['activo']           # eliminar por clave
print('Después de eliminar activo:', producto)
````

````python
# Verificar si una clave existe
print('¿Existe precio?:', 'precio' in producto)
print('¿Existe color?: ', 'color' in producto)
````

## Iterando sobre un Diccionario

Muy útil para generar reportes automáticos.

````python
# Iterar sobre claves y valores con .items()
empleado = {
    'nombre': 'María González',
    'id': 'E-2045',
    'departamento': 'Finanzas',
    'salario': 4200,
    'ciudad': 'Quito'
}

print('=== FICHA DE EMPLEADO ===')
for clave, valor in empleado.items():
    print(f'  {clave.capitalize():15}: {valor}')
````

````python
# Diccionario de diccionarios — como una tabla de base de datos
inventario = {
    'PROD-001': {'nombre': 'Laptop', 'precio': 899, 'stock': 10},
    'PROD-002': {'nombre': 'Mouse',  'precio': 25,  'stock': 150},
    'PROD-003': {'nombre': 'Monitor','precio': 350, 'stock': 25}
}

print('=== INVENTARIO ===')
for codigo, datos in inventario.items():
    valor_total = datos['precio'] * datos['stock']
    print(f"{codigo} | {datos['nombre']:10} | ${datos['precio']:>5} | Stock: {datos['stock']:>3} | Valor: ${valor_total:,}")
````

## ✅ Resumen del Módulo 6

- Un **diccionario** almacena pares `clave: valor` `{clave: valor, ...}`
- Acceso: `dict['clave']` o `dict.get('clave', default)`
- Métodos: `.keys()`, `.values()`, `.items()`
- Agregar/modificar: `dict['nueva_clave'] = valor`
- Eliminar: `del dict['clave']`
- Iterar con `for clave, valor in dict.items()`
- Los diccionarios son la base de JSON y muchos formatos de datos modernos
