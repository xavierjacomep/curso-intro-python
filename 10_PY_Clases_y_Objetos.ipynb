## ¿Qué es la Programación Orientada a Objetos?

La **Programación Orientada a Objetos (POO)** organiza el código alrededor de **objetos** que combinan **datos** (atributos) y **comportamiento** (métodos).

**Analogía:** Piensa en un producto en tu empresa:
- **Datos (atributos):** nombre, precio, stock, categoría
- **Comportamiento (métodos):** calcular precio con descuento, verificar disponibilidad, actualizar stock

Una **clase** es el molde o plantilla. Un **objeto** es una instancia concreta de esa clase.

## Crear una Clase

ChatGPT Image 13 may 2026, 03_54_01 p.m..png

````python
# Definir una clase Producto
class Producto:
    """
    Clase para representar un producto del inventario.

    Atributos:
        nombre (str): Nombre del producto
        precio (float): Precio base
        stock (int): Unidades disponibles
        categoria (str): Categoría del producto
    """

    # __init__ es el constructor: se ejecuta al crear cada objeto
    def __init__(self, nombre, precio, stock, categoria='General'):
        self.nombre    = nombre
        self.precio    = precio
        self.stock     = stock
        self.categoria = categoria

    # Método: calcular precio con descuento
    def precio_con_descuento(self, porcentaje):
        """Retorna el precio con un descuento aplicado."""
        return self.precio * (1 - porcentaje / 100)

    # Método: verificar disponibilidad
    def esta_disponible(self):
        """Retorna True si hay stock."""
        return self.stock > 0

    # Método: actualizar stock
    def vender(self, cantidad):
        """Reduce el stock al vender una cantidad."""
        if cantidad > self.stock:
            print(f'⚠️ Stock insuficiente. Disponible: {self.stock}')
        else:
            self.stock -= cantidad
            print(f'Venta registrada. Stock restante: {self.stock}')

    # Método especial __str__: define cómo se muestra el objeto con print()
    def __str__(self):
        return f'[{self.categoria}] {self.nombre} - ${self.precio:.2f} (Stock: {self.stock})'
````

````python
# Crear instancias (objetos) de la clase Producto
laptop  = Producto('Laptop ProX', 899.99, 25, 'Electrónica')
mouse   = Producto('Mouse Inalámbrico', 35.00, 150, 'Accesorios')
silla   = Producto('Silla Ergonómica', 280.00, 0, 'Muebles')  # sin stock

# print() usa el método __str__
print(laptop)
print(mouse)
print(silla)
````

````python
# Acceder a atributos con notación punto
print(f'Nombre: {laptop.nombre}')
print(f'Precio: ${laptop.precio}')
print(f'Stock:  {laptop.stock}')
````

````python
# Llamar métodos
print('Precio con 15% descuento:', laptop.precio_con_descuento(15))
print('¿Laptop disponible?:', laptop.esta_disponible())
print('¿Silla disponible?: ', silla.esta_disponible())
````

````python
# Usar el método vender
laptop.vender(5)   # vende 5 unidades
laptop.vender(30)  # intenta vender más de lo disponible
````

````python
# Modificar atributos directamente
laptop.precio = 849.99
print(f'Nuevo precio: ${laptop.precio}')
````

## Ejemplo completo: clase Empleado

````python
class Empleado:
    """Representa un empleado de la empresa."""

    # Atributo de clase (compartido por todos los objetos)
    empresa = 'Corporación Andina S.A.'

    def __init__(self, nombre, departamento, salario):
        self.nombre       = nombre
        self.departamento = departamento
        self.salario      = salario
        self.bonos        = []

    def agregar_bono(self, monto):
        """Agrega un bono al historial."""
        self.bonos.append(monto)
        print(f'Bono de ${monto:,} registrado para {self.nombre}')

    def salario_total(self):
        """Calcula salario + total de bonos."""
        return self.salario + sum(self.bonos)

    def reporte(self):
        """Imprime ficha del empleado."""
        print(f'=== {self.empresa} ===')
        print(f'Empleado:     {self.nombre}')
        print(f'Departamento: {self.departamento}')
        print(f'Salario base: ${self.salario:,}')
        print(f'Bonos:        ${sum(self.bonos):,}')
        print(f'Total:        ${self.salario_total():,}')

# Crear empleados
ana  = Empleado('Ana García', 'Finanzas', 4000)
luis = Empleado('Luis Pérez', 'Ventas', 3200)

# Agregar bonos
ana.agregar_bono(500)
ana.agregar_bono(300)
luis.agregar_bono(800)

print()
ana.reporte()
print()
luis.reporte()
````

class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo
    
    def depositar(self, monto):
        self.saldo += monto
        print(f'Depósito: ${monto:,} | Saldo: ${self.saldo:,}')
    
    def retirar(self, monto):
        if monto > self.saldo:
            print('Saldo insuficiente')
        else:
            self.saldo -= monto
            print(f'Retiro: ${monto:,} | Saldo: ${self.saldo:,}')
    
    def mostrar_saldo(self):
        print(f'{self.titular} - Saldo: ${self.saldo:,}')

cuenta = CuentaBancaria('María López', 1000)
cuenta.depositar(500)
cuenta.retirar(200)
cuenta.mostrar_saldo()

## ✅ Resumen del Módulo 10

- Una **clase** es un molde: define atributos y métodos
- Un **objeto** es una instancia concreta de una clase
- **`__init__`**: constructor que inicializa el objeto

- **Atributos**: datos del objeto (`self.nombre`)
- **Métodos**: funciones del objeto (`def calcular(self):`)
- **`__str__`**: define cómo se muestra el objeto con `print()`
