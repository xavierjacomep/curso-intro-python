## ¿Qué es un Conjunto?

Un **conjunto (set)** es una colección **sin orden** y **sin duplicados**. Python elimina automáticamente los elementos repetidos.

**Analogía empresarial:** Imagina que tienes una lista de clientes que compraron distintos productos. Un conjunto te dice **qué clientes únicos** compraron, sin importar cuántas veces aparezca el mismo.

Se crean con llaves `{}` o con `set()`.

---
<hr>

````python
# Crear un conjunto — los duplicados se eliminan automáticamente
generos = {'Rock', 'Pop', 'Jazz', 'Rock', 'Pop', 'Blues'}
print(generos)  # Rock y Pop aparecen solo una vez
````

````python
# Ejemplo práctico: obtener categorías únicas de una lista de productos
categorias_lista = ['Electrónica', 'Ropa', 'Electrónica', 'Alimentos',
                    'Ropa', 'Electrónica', 'Muebles', 'Alimentos']

categorias_unicas = set(categorias_lista)
print('Categorías únicas:', categorias_unicas)
print('Total de categorías distintas:', len(categorias_unicas))
````

````python
# Agregar elementos con add()
ciudades_clientes = {'Quito', 'Guayaquil', 'Cuenca'}
ciudades_clientes.add('Loja')
ciudades_clientes.add('Quito')  # ya existe, no se agrega
print(ciudades_clientes)
````

````python
# Eliminar con remove()
ciudades_clientes.remove('Loja')
print(ciudades_clientes)
````

````python
# Verificar si un elemento está en el conjunto
print('¿Quito está en el conjunto?:', 'Quito' in ciudades_clientes)
print('¿Ambato está?:              ', 'Ambato' in ciudades_clientes)
````

## Operaciones de Teoría de Conjuntos

Estas operaciones son muy útiles en análisis de datos para comparar grupos:

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/SetsLogic.png" width="650">

````python
# Escenario: clientes que compraron en dos campañas diferentes
clientes_campana_A = {'Ana', 'Luis', 'María', 'Carlos', 'Pedro'}
clientes_campana_B = {'Carlos', 'Pedro', 'Sofía', 'Diego', 'Ana'}

# Intersección: clientes que compraron en AMBAS campañas (&)
compraron_ambas = clientes_campana_A & clientes_campana_B
print('Compraron en ambas campañas:', compraron_ambas)

# Unión: todos los clientes únicos (|)
todos_clientes = clientes_campana_A | clientes_campana_B
print('Total de clientes únicos:   ', todos_clientes)

# Diferencia: clientes de A que NO compraron en B (-)
solo_A = clientes_campana_A - clientes_campana_B
print('Solo compraron en campaña A:', solo_A)

solo_B = clientes_campana_B - clientes_campana_A
print('Solo compraron en campaña B:', solo_B)
````

````python
# También se pueden usar métodos
print('Intersección:', clientes_campana_A.intersection(clientes_campana_B))
print('Unión:       ', clientes_campana_A.union(clientes_campana_B))
print('Diferencia:  ', clientes_campana_A.difference(clientes_campana_B))
````

````python
# Subconjunto e Superconjunto
clientes_vip = {'Carlos', 'Ana'}
todos = {'Ana', 'Luis', 'María', 'Carlos', 'Pedro'}

print('¿VIP es subconjunto de todos?:', clientes_vip.issubset(todos))
print('¿Todos es superconjunto de VIP?:', todos.issuperset(clientes_vip))
````
